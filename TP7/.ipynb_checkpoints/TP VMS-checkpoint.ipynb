Dans ce TP, nous aimerions explorer le classifieur machineà vecteurs de support. Nous essayerons d'utiliser plusieurs noyaux et voir la difference entre leur capacité de séparation.

# Importation des librairies necessaires au travail

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.io import loadmat
from sklearn.svm import SVC
```

# Lecture des fichiers de données
Pour ce TP, nous allons lire les données à partir d'un fichier mat.

```python
mat = loadmat("data.mat")
X = mat["X"]
y = mat["y"]
X.shape
```

# Plotting data

```python
def plotData(X,y):
    m,n = X.shape[0],X.shape[1]
    pos2,neg2= (y==1).reshape(m,1), (y==0).reshape(m,1)
    plt.figure(figsize=(8,6))
    plt.scatter(X[pos2[:,0],0],X[pos2[:,0],1],c="black",marker="+")
    plt.scatter(X[neg2[:,0],0],X[neg2[:,0],1],c="y",marker="o")
    plt.xlim(0,1)
    plt.ylim(0.4,1)
plotData(X,y)
```

```python
def plotBoudary(X,y, classifier, type_):
    # plotting the decision boundary
    X_5,X_6 = np.meshgrid(np.linspace(X[:,0].min(),X[:,1].max(),num=100),np.linspace(X[:,1].min(),X[:,1].max(),num=100))
    plt.contour(X_5,X_6,classifier.predict(np.array([X_5.ravel(),X_6.ravel()]).T).reshape(X_5.shape),1,colors="b")
    plt.title('Decision boundary with '+ type_+' Kernel')
    plt.xlim(0,1)
    plt.ylim(0.4,1)
```

# Entrainement des données avec plusieurs noyaux

```python
classifier1 = SVC(kernel="linear",gamma=30)
classifier1.fit(X,y.ravel())

classifier2 = SVC(kernel="sigmoid",gamma=30)
classifier2.fit(X,y.ravel())


classifier3 = SVC(kernel="rbf",gamma=30)
classifier3.fit(X,y.ravel())

classifier4 = SVC(kernel="poly",gamma=30)
classifier4.fit(X,y.ravel())

```

# Vecteurs de support et séparation de données

```python
#Decision boundary with linear kernel
plotData(X,y)
plotBoudary(X,y,classifier1,"linear")
#Decision boundary with sigmoid kernel
plotData(X,y)
plotBoudary(X,y,classifier2,"sigmoid")
#Decision boundary with RBF kernel
plotData(X,y)
plotBoudary(X,y,classifier3,"RBF")
#Decision boundary with polynomial kernel
plotData(X,y)
plotBoudary(X,y,classifier4,"polynomial")

```
